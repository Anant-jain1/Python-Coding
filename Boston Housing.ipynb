```python
quantiles_CRIM = house['CRIM'].quantile([0.25,0.5,0.75])
ub_CRIM = quantiles_CRIM[0.75]+1.5*(quantiles_CRIM[0.75]-quantiles_CRIM[0.25])
lb_CRIM = quantiles_CRIM[0.25]-1.5*(quantiles_CRIM[0.75]-quantiles_CRIM[0.25])
house['CRIM']=np.where(house['CRIM']>ub_CRIM,house['CRIM'].median(),house['CRIM'])
house['CRIM']=np.where(house['CRIM']<lb_CRIM,house['CRIM'].median(),house['CRIM'])
```

```python
quantiles_B = house['B'].quantile([0.25,0.5,0.75])
ub_B = quantiles_B[0.75]+1.5*(quantiles_B[0.75]-quantiles_B[0.25])
lb_B = quantiles_B[0.25]-1.5*(quantiles_B[0.75]-quantiles_B[0.25])
house['B']=np.where(house['B']>ub_B,house['B'].median(),house['B'])
house['B']=np.where(house['B']<lb_B,house['B'].median(),house['B'])

quantiles_MEDV = house['MEDV'].quantile([0.25,0.5,0.75])
ub_MEDV = quantiles_MEDV[0.75]+1.5*(quantiles_MEDV[0.75]-quantiles_MEDV[0.25])
lb_MEDV = quantiles_MEDV[0.25]-1.5*(quantiles_MEDV[0.75]-quantiles_MEDV[0.25])
house['MEDV']=np.where(house['MEDV']>ub_MEDV,house['MEDV'].median(),house['MEDV'])
house['MEDV']=np.where(house['MEDV']<lb_MEDV,house['MEDV'].median(),house['MEDV'])
```

```python
def getCorrelatedFeature(corrdata, threshold):
    feature=[]
    value=[]
    
    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index])> threshold:
            feature.append(index)
            value.append(corrdata[index])
            
    df = pd.DataFrame(data = value,index = feature, columns=['Corr Value'])
    return df
```

```python
# We first take a threshold value of 0.50 and form the model
threshold = 0.50
corr_value= getCorrelatedFeature(corrmat['MEDV'],threshold)
corr_value
```

```python
correlated_data= house[corr_value.index]
correlated_data.head()
```

```python
sns.pairplot(correlated_data)
plt.tight_layout()
```

```python
sns.heatmap(correlated_data.corr(), annot=True, annot_kws={'size':12})
```

```python
X = correlated_data.drop(labels=['MEDV'],axis=1)
Y = correlated_data['MEDV']
X.head
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,train_size=0.7,random_state=100)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_predict = model.predict(X_test)
df = pd.DataFrame( data =[Y_predict, Y_test])
df.T
```

```python
from sklearn.metrics import mean_squared_error , r2_score
mse = mean_squared_error(Y_test,Y_predict)
mse
```

```python
r_squared = r2_score(Y_test,Y_predict)
r_squared
```
